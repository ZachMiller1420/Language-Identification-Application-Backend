### Create a fun to count and remove punctuation from the Text

```python
all_punctuation = string.punctuation

def count_punctuation(text):
    count = sum(1 for char in text if char in all_punctuation)
    return count

total_punctuation_count = df['Text'].apply(count_punctuation).sum()

print(total_punctuation_count)
```

```python
def remove_pun(text):
    for pun in string.punctuation:
        text = text.replace(pun, "")
    text = text.lower()
    return text

df['Text'] = df['Text'].apply(remove_pun)

updated_punctuation_count = df['Text'].apply(remove_pun).sum()
```

```python
# Remove integers from 'Text' feature
df['Text'] = df['Text'].apply(lambda x: ''.join([i for i in x if not i.isdigit()]))
```

## Average Text length by Language

In this visualization, we examine the average length of text for each unique language feature. It's evident that most languages exhibit similar text lengths, with one notable exception: Hindi. Hindi text stands out for its significantly longer average length compared to other languages in the dataset.

```python
# Calculate the length of each 'Text' entry
df['Text Length'] = df['Text'].apply(lambda x: len(x.split()))

# Group by 'Language' and calculate the mean of 'Text Length' for each group
average_length_by_language = df.groupby('Language')['Text Length'].mean()

colors = ['blue', 'green', 'red', 'yellow', 'lightskyblue', 'lightcoral', 'magenta', 'cyan', 'lightgreen', 'purple', 'lightpink', 'lightblue', 'lime', 'lightgray', 'brown', 'pink', 'black']
# Plot the results in a histogram
plt.figure(figsize=(5, 3))
plt.bar(average_length_by_language.index, average_length_by_language, color=colors)
plt.xlabel('Language')
plt.ylabel('Average Length of Text')
plt.title('Average Length of Text by Language')
plt.xticks(rotation=90)
plt.show()
```

## Word Clouds
The word clouds displayed in the visualizations provide a graphical representation of the most frequently occurring words in the text data for each language category. Each word cloud contains words from the respective language, with the size of each word proportional to its frequency in the text data.

The benefit of using word clouds in this context lies in their ability to quickly convey the most common words within each language category. By visually representing word frequency, word clouds offer insights into the prominent themes, topics, or vocabulary associated with different languages. This can aid in understanding the linguistic characteristics and nuances specific to each language, which can be valuable for tasks such as language identification, sentiment analysis, or topic modeling.

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define font file paths for each language
font_files = {
    'Malayalam': 'NotoSansMalayalam-VariableFont_wdth,wght.ttf',
    'Arabic': 'NotoKufiArabic-VariableFont_wght.ttf',
    'Tamil': 'NotoSansTamil-VariableFont_wdth,wght.ttf',
    'Hindi': 'TiroDevanagariHindi-Regular.ttf',
    'Kannada': 'NotoSansKannada-VariableFont_wdth,wght.ttf'
}

# Define a function to generate word clouds for each language
def generate_word_cloud(text, language):
    # Combine all text samples for the given language
    text_combined = ' '.join(text)
    # Generate word cloud
    font_file = font_files.get(language, None)  # Default to a regular font
    wordcloud = WordCloud(width=800, height=400, colormap="Dark2", font_path=font_file).generate(text_combined)
    # Display word cloud
    plt.figure(figsize=(5, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.title(language)
    plt.axis('off')
    plt.show()

# Group the data by language and apply the function to each group
for language, group in df.groupby('Language'):
    generate_word_cloud(group['Text'], language)


```

## Train Test Split the Data set
In this section, we're preparing our dataset for machine learning by splitting it into training and testing sets using the train_test_split function from the sklearn.model_selection module. Our dataset consists of features (X) and the corresponding target variable (Y). We allocate 20% of the data for testing (test_size=.2), ensuring that our model will be evaluated on unseen data. 

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0]
Y = df.iloc[:,1]

X_train, X_test, y_train, y_test = train_test_split( X, Y, test_size=.2)
```

Next, we initialize a TfidfVectorizer from sklearn.feature_extraction.text, which converts our text data into numerical vectors. We configure the vectorizer to consider both single words (unigrams) and pairs of consecutive words (bigrams) as features (ngram_range=(1,2)), and to use character-level n-grams for feature extraction (analyzer='char'). This step is crucial as it transforms our raw text data into a format that machine learning algorithms can understand, facilitating the subsequent training of our model.

```python
from sklearn import feature_extraction

vec = feature_extraction.text.TfidfVectorizer(ngram_range=(1,2), analyzer='char')
```

### Visualizing the learning curve
Visualizing a learning curve, which provides insights into the performance of a machine learning model, particularly its ability to generalize to unseen data. The plot_learning_curve function defined here takes various parameters such as the estimator (the machine learning model), the title for the plot, the training data (X and y), and optional parameters like the number of cross-validation folds (cv) and the range of training sizes (train_sizes). The function then utilizes the learning_curve function from scikit-learn to compute the training and cross-validation scores for different training set sizes. These scores are then plotted against the number of training examples, with shaded regions representing the variance of the scores. By examining how the training and cross-validation scores change with the number of training examples, one can assess the model's bias-variance tradeoff and determine if the model is underfitting, overfitting, or performing optimally. This visualization aids in making informed decisions about model complexity, dataset size, and potential improvements to achieve better generalization performance. In this specific example, the learning curve is plotted for a logistic regression model trained on the provided training data (X_train and y_train).


```python
from sklearn.model_selection import learning_curve

# Define function to plot learning curve
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
 
    plt.figure(figsize=(10, 6))
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    plt.tight_layout()  # Adjust layout to prevent overlap
    return plt

# Plot learning curve for logistic regression model
plot_learning_curve(lr_model_pipeline, "Learning Curve (Logistic Regression)", X_train, y_train, cv=5)
plt.show()
```

# Support Vector Machines (SVM)

Now lets take a look at the SVM algorithm! The Support Vector Machine algorithm is well-suited for Natural Language Processing tasks due to its ability to effectively handle high-dimensional data, such as the large number of features generated from text data. SVMs are robust to overfitting and can capture complex patterns in text data through kernel functions. They perform particularly well in scenarios with a small number of samples compared to the feature space, which is common in NLP applications. Overall, SVMs offer a versatile and reliable approach for various NLP tasks, including text classification and sentiment analysis.

In the context of the learning curve graph, a significant gap between the training and validation scores typically indicates overfitting. The wider the gap, the higher the likelihood of overfitting.
